```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
avocado_df = pd.read_csv('avocado.csv')
avocado_df.head()
```

```python
avocado_df.info()
```

```python
avocado_df['Date'] = pd.to_datetime(avocado_df['Date'])
avocado_df.info()
```

```python
filtered_df = avocado_df.query(
    "year == 2015 & type == 'conventional' & region == 'Albany'"
)[['Date', 'Total Bags', 'Small Bags']]
filtered_df.head()
```

```python
melted_df = pd.melt(
    filtered_df,
    id_vars='Date',
    value_vars=['Total Bags', 'Small Bags'],
    var_name='Bags',
    value_name='Count'
)
melted_df.head()
```

```python
sns.lineplot(data=melted_df, x='Date', y='Count', hue='Bags', errorbar='sd')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Amount of Bags Over Time')
```

```python
bins = pd.qcut(avocado_df['Total Volume'], 4, labels=['Poor', 'Modest', 'Good', 'Excellent'])
avocado_df['Sales Volume'] = bins
avocado_df.head()
```

```python
sns.countplot(data=avocado_df, x='year', hue='Sales Volume')
plt.title('Yearly Sales Volume')
```

```python
melted_bags_df = pd.melt(
    avocado_df,
    id_vars='year',
    value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'],
    var_name='Bags',
    value_name='Count'
)
melted_bags_df.head()
```

```python
sns.relplot(
    data=melted_bags_df,
    kind='line',
    x='year',
    y='Count',
    hue='Bags'
)
plt.title('Sizes Sold Over Years')
```
